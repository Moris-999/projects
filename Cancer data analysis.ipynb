```python
### checking for outliers
outliers= Data[numerical_columns].apply(lambda x:(x - x.mean()).abs()>3*x.std()).sum()

print ("\nOutliers:")
print(outliers)
```

```python
 ### Define the z-score threshold for outlier detection
threshold=3

###Caculate z-scores for each numerical column
z_scores=Data[numerical_columns].apply(lambda x:(x-x.mean()) /x.std())

###Find rows where any z-score exceeds the threshold
outlier_rows=z_scores[(z_scores.abs()>threshold).any (axis=1)]
outlier_rows

```

```python
### removing outlier from the dataframe
Data_flitered=Data.drop(outlier_rows.index)

## Display flitred dataframe
print (Data_flitered)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Select the columns for the heatmap
development_risk = Data_flitered[['Cancer Type', 'Male Risk Development Percentage', 'Female Risk Development Percentage']]
dying_risk = Data_flitered[['Cancer Type', 'Male Risk Dying Percentage', 'Female Risk Dying Percentage']]

# Set the index to 'Cancer Type'
development_risk.set_index('Cancer Type', inplace=True)
dying_risk.set_index('Cancer Type', inplace=True)

# Create a function to plot a heatmap using pandas
def plot_heatmap(data, title):
    plt.figure(figsize=(12, 6))
    plt.imshow(data, cmap='coolwarm', aspect='auto')
    
    plt.colorbar(label='Percentage')
    
    plt.xticks(range(data.shape[1]), data.columns, rotation=45, ha='right')
    plt.yticks(range(data.shape[0]), data.index)
    
    plt.title(title)
    plt.xlabel('Gender')
    plt.ylabel('Cancer Type')

    # Annotate the heatmap with the actual values
    for i in range(data.shape[0]):
        for j in range(data.shape[1]):
            plt.text(j, i, f'{data.iloc[i, j]:.1f}', ha='center', va='center', color='black')

    plt.tight_layout()
    plt.show()

# Plotting the heatmaps
plot_heatmap(development_risk, 'Heatmap of Development Risk Percentage (Male vs Female)')
plot_heatmap(dying_risk, 'Heatmap of Dying Risk Percentage (Male vs Female)')
```

The first heatmap show male risk development have a higher chance of developing prostate cancer 12.9% than female there dont have any case reported.
The second heatmap shows dying risk percentage male stand a higher chance developing lung and bronchus 44%  than female 38%.
